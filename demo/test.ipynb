Now, InterSystems IRIS supports vectors as a datatype in tables! Here, we create a table with a few different columns. The last column, 'description_vector', will be used to store vectors that are generated by passing the 'description' of a review through an embedding model.

```python

result = cursor.execute("Create TABLE Test.Demo (vec1 VECTOR)")
print(result)
```

```python
result = cursor.execute("INSERT INTO Test.Demo (vec1) VALUES ('1,2,3')")
result = cursor.execute("INSERT INTO Test.Demo (vec1) VALUES ('2,2,1')")
result = cursor.execute("INSERT INTO Test.Demo (vec1) VALUES ('1,1,4')")
```

```python
cursor.execute("SELECT TOP 3 * FROM Test.Demo ORDER BY VECTOR_COSINE(vec1, TO_VECTOR('2,4,6', DOUBLE)) DESC")
results = cursor.fetchall()
```
